## Вариант 2

*Условие:* Пусть $X_1 \dots X_n$ – i.i.d. величины из некоторого распределения, которое параметризуется параметром $\theta \in E$. Пусть $\overline{X}_n=(X_1+\dots+X_n)/n;\ \mu_\theta$ – математическое ожидание случайной величины $X_i$. Найти с помощью неравенства Чебышева и центральной предельной теорему номер $n_{\delta,\varepsilon}$, при котором для любого $\theta \in E$ и заданных $\delta,\ \varepsilon \in (0,1)$ выполняется соотношение

$$
P(|\overline{X}_{n_{\delta,\varepsilon}}-\mu_\theta|\le \varepsilon) \ge 1 - \delta
$$

Сгенерировать 100 выборок найденного с помощью неравенства Чебвшева объема $n_{\delta,\varepsilon}$ и посчитать количество и долю выборок, для которых $|\overline{X}_{\delta\varepsilon}-\mu_\theta|\le \varepsilon$. Параметры эксперимента: $\varepsilon=0.01,\ \delta=0.05$ и $\theta$, при котором найдено $n_{\delta,\varepsilon}$. То же самое сделать и для $n_{\delta,\varepsilon}$, найденного с помощью ЦПТ

Распределение: $Pois(\lambda), \lambda \in E = (0,10)$

### Неравенство Чебышева

Для распределений, заданных параметром $\theta$:
$$
P(|\overline{X}_{n_{\delta,\varepsilon}}-\mu_\theta|\le \varepsilon) \ge 1 - \frac{\sigma^2_\theta}{n \varepsilon^2}
$$

Следовательно $\delta=\frac{\sigma^2_\theta}{n \varepsilon^2}$. Следовательно $n=\frac{\sigma^2_\theta}{\delta \varepsilon^2}$. Чтобы работало для всех $\theta \in (0, 10)$ получим:
$$
n=\frac{\sup \sigma^2_\theta}{\delta \varepsilon^2}
$$

В распределение Пуассона с параметрами $\theta \in (0,10)$ максимальной дисперсией будет $10$ при $\theta=10$. Следовательно:

$$
n=\frac{10}{\delta \varepsilon^2}
$$

### ЦПТ

$$
\frac{\sqrt{n}(\overline{X}_{n_{\delta,\varepsilon}} - \mu_\theta)}{\sigma_\theta} \to Z,\ Z \sim N(0,1)
$$

Подставив в изначальное выражение, получим:
$$
P(|Z| \le \frac{\varepsilon \sqrt{n}}{\sigma_\theta}) \ge 1 - \delta
$$

Воспользовавшись стандартными значениями $Z$ получим, что:
$$
\frac{\varepsilon \sqrt{n}}{\sigma_\theta} \ge z_{1-\delta/2}
$$

Следовательно:
$$\sqrt{n}\ge \frac {z_{1-\delta/2} \sigma_\theta}{\varepsilon}$$

$$n \ge \frac{z_{1-\delta/2}^2 \sigma^2_\theta}{\varepsilon^2}$$

Чтобы оно работало для всех $\theta \in (0,10)$:

$$
n \ge \frac{z^2_{1-\delta/2} \cdot 10}{\varepsilon^2}
$$

```python
import numpy as np
from scipy.stats import norm
```

Задаем параметры:

```python
delta = 0.05
epsilon = 0.01
```

$n_{\delta,\varepsilon}$ найдено при $\theta=10$ при обоих методах, следовательно выбираем его параметром

```python
theta = 10
```

```python
n_sim = 100
```

```python
n_cheb = round(theta / (delta * epsilon ** 2))
print(f"Найденное n с помощью нерасенства Чебышева: {n_cheb}")
```

```python
z = norm.ppf(1 - delta / 2)
print(z)
```

```python
n_cpt = round(z ** 2 * theta / epsilon ** 2)
print(f"Найденное n с помощью нерасенства ЦПТ: {n_cpt}")
```

```python
samples_cheb = np.random.poisson(theta, (n_sim, n_cheb))
samples_cpt = np.random.poisson(theta, (n_sim, n_cpt))

means_cheb = samples_cheb.mean(axis=1)
means_cpt = samples_cpt.mean(axis=1)

success_cheb = np.mean(np.abs(means_cheb - theta) <= epsilon)
success_cpt = np.mean(np.abs(means_cpt - theta) <= epsilon)

print(f"Доля выборок (Чебышев): {success_cheb}")
print(f"Доля выборок (ЦПТ): {success_cpt}")
```
